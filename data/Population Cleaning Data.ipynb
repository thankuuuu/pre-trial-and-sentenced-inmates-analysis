```python
#Clean Data 
import pandas as pd 
import numpy as np

import warnings
warnings.filterwarnings('ignore')

#import the pretrial-dataset
Population_Gender = pd.read_csv('ACSDT1Y2018.B01001_data_with_overlays_2020-03-21T150845.csv', header = 1)
Population_Race = pd.read_csv('ACSDT1Y2018.B03002_data_with_overlays_2020-03-30T145930.csv', header = 1)
#Check whether any datasets have null or not
Population_Gender.info()
Population_Race.info() 
```

```python
Population_Gender.head()
```

```python
Population_Race.head()
```

```python
#Population by Gender Cleaning
#Selecting just only used columns
Population_Gender2 = Population_Gender[['Geographic Area Name','Estimate!!Total','Estimate!!Total!!Male','Estimate!!Total!!Female']]
Population_Gender2 = Population_Gender2.set_index('Geographic Area Name')
```

```python
Population_Gender2.head()
```

```python
#Sum up all states data
Sum_Gender = Population_Gender2.sum(axis=0).to_frame()
Sum_Gender
```

```python
#Population by Gender Cleaning
#Selecting just only used columns
Population_Race2 = Population_Race.iloc[:,[1,6,8,10,12,14,16,18,24]]
Population_Race2 = Population_Race2.set_index('Geographic Area Name')
```

```python
Population_Race2.head()
```

```python
#Sum up all states data
Sum_Race = Population_Race2.sum(axis=0).to_frame()
Sum_Race
```

```python
Sum = pd.concat([Sum_Gender,Sum_Race])
Sum
```

```python
Sum = Sum.rename(index={'Estimate!!Total': 'Total',
                        'Estimate!!Total!!Male':'Male',
                        'Estimate!!Total!!Female':'Female',
                        'Estimate!!Total!!Not Hispanic or Latino!!White alone':'White',
                        'Estimate!!Total!!Not Hispanic or Latino!!Black or African American alone':'Black',
                        'Estimate!!Total!!Not Hispanic or Latino!!American Indian and Alaska Native alone':'American IND',
                        'Estimate!!Total!!Not Hispanic or Latino!!Asian alone':'Asian',
                        'Estimate!!Total!!Not Hispanic or Latino!!Native Hawaiian and Other Pacific Islander alone':'Hawaiian',
                        'Estimate!!Total!!Not Hispanic or Latino!!Some other race alone':'Other',
                        'Estimate!!Total!!Not Hispanic or Latino!!Two or more races':'TwoRace',
                        'Estimate!!Total!!Hispanic or Latino':'Hispanic',
                       })
```

```python
Sum.iloc[8] = Sum.iloc[8]+Sum.iloc[7]+Sum.iloc[9] 
Sum = Sum.drop(Sum.index[7])
Sum = Sum.drop(Sum.index[8])
```

```python
Sum = Sum.rename(columns = {0:'Number'})
Sum
```

```python
# Export as csv file
Sum.to_csv('Population2018.csv')
```
