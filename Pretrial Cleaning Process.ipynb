```python
#Clean Data 
import pandas as pd 
import numpy as np

import warnings
warnings.filterwarnings('ignore')

#import the pretrial-dataset
Pretrial = pd.read_csv('Accused_Pre-Trial_Inmates_in_Correctional_Facilities.csv')
#Check whether any datasets have null or not
Pretrial.shape
```

```python
Pretrial.head()
```

```python
#import the sentenced inmates dataset
Sentenced_Inmates = pd.read_csv('Sentenced_Inmates_in_Correctional_Facilities.csv')
#Check whether any datasets have null or not
Sentenced_Inmates.shape
```

```python
Sentenced_Inmates.head()
```

```python
# Extracting Year of pre-trail data
year = []
for i in range (len(Pretrial.index)):
        year.append((Pretrial.loc[i,'LATEST ADMISSION DATE'][6:10]))

join = {'Year':year}
df2 = pd.DataFrame(join)
Data_pretrial = pd.concat([Pretrial,df2], axis=1)
Data_pretrial.shape
```

```python
# Extracting Year of sentenced inmates data
year = []
for i in range (len(Sentenced_Inmates.index)):
        year.append((Sentenced_Inmates.loc[i,'LATEST ADMISSION DATE'][6:10]))

join = {'Year':year}
df2 = pd.DataFrame(join)
Data_sentenced = pd.concat([Sentenced_Inmates,df2], axis=1)
Data_sentenced.shape
```

```python
#Selecting just only 2018 data
Data_pretrial = Data_pretrial.loc[Data_pretrial.Year == '2018']
Data_sentenced = Data_sentenced.loc[Data_sentenced.Year == '2018']
```

```python
#Copy dataframe for searching sentence day after found all inmates who sentenced after pre-trial process.
Data_pretrial_copy = Data_pretrial.copy()
Data_sentenced_copy = Data_sentenced.copy()
```

```python
Data_pretrial.shape
```

```python
Data_pretrial.columns
```

```python
#Deleting unsued columns
Data_pretrial = Data_pretrial.drop(columns=['BOND AMOUNT','DOWNLOAD DATE','FACILITY','AGE','DETAINER                                        '])
```

```python
#Deleting Duplicate indentifier as they uploaded all inmates data everyday so it contains many duplicates data
Data_pretrial = Data_pretrial.drop_duplicates()
Data_pretrial.shape
```

```python
Data_sentenced.shape
```

```python
Data_sentenced.columns
```

```python
#Deleting unsued columns
Data_sentenced = Data_sentenced.drop(columns=['DOWNLOAD DATE','END SENTENCE DATE','FACILITY','DETAINER','AGE','SENTENCE DAYS','SPECIAL PAROLE END DATE                                                        '])
```

```python
#Deleting Duplicate indentifier as they uploaded all inmates data everyday so it contains many duplicates data
Data_sentenced = Data_sentenced.drop_duplicates()
Data_sentenced.shape
```

```python
Sentenced_Pretrial= pd.merge(Data_pretrial, Data_sentenced, how='inner')
Sentenced_Pretrial.head()
```

```python
Sentenced_Pretrial.shape
```

```python
#Identify sentenced date
Data_sentenced_copy = Data_sentenced_copy.drop(columns=['DOWNLOAD DATE','END SENTENCE DATE','AGE','FACILITY','DETAINER','SPECIAL PAROLE END DATE                                                        '])
Data_pretrial_copy = Data_pretrial_copy.drop(columns=['DOWNLOAD DATE','FACILITY','AGE','DETAINER                                        '])
```

```python
Data_sentenced_copy = Data_sentenced_copy.rename(columns={'SENTENCE DAYS':'SENTENCE_DAYS'})
```

```python
Data_sentenced_group = Data_sentenced_copy.groupby(['IDENTIFIER','LATEST ADMISSION DATE','RACE','GENDER','OFFENSE'], as_index=False).SENTENCE_DAYS.max()
```

```python
#Joining Data_pretrial and Data_sentenced dataframes with the same identifier to see all people who were sentenced after pre-trial
Sentenced_Pretria_sentenced_date = pd.merge(Sentenced_Pretrial, Data_sentenced_group, how='inner')
Sentenced_Pretria_sentenced_date 
```

```python
Data_pretrial_copy = Data_pretrial_copy.rename(columns={'BOND AMOUNT':'BOND_AMOUNT'})
```

```python
Data_pretrial_group = Data_pretrial_copy.groupby(['IDENTIFIER','LATEST ADMISSION DATE','RACE','GENDER','OFFENSE'], as_index=False).BOND_AMOUNT.max()
```

```python
#Joining Data_pretrial and Data_sentenced dataframes with the same identifier to see all people who were sentenced after pre-trial
Sentenced_Pretria_sentenced_date2 = pd.merge(Sentenced_Pretria_sentenced_date, Data_pretrial_group, how='inner')
Sentenced_Pretria_sentenced_date2
```

```python
#Find top 10 offense
top_offense = Sentenced_Pretria_sentenced_date2.groupby(['OFFENSE'], as_index=False)
top_offense_sum  = top_offense.size()
top_offense_sum = top_offense_sum.to_frame()
top_offense_sum.sort_values([0], axis=0,ascending=False, inplace=True) 
top_offense_sum.head(10)
```

```python
#Change others offense beside the top 10 as others
#Drop all index
Sentenced_Pretria_sentenced_date2.reset_index(drop=True, inplace=True)
Data_pretrial.reset_index(drop=True, inplace=True)
Data_sentenced.reset_index(drop=True, inplace=True)
Sentenced_Pretrial.reset_index(drop=True, inplace=True)

for i in range(len(Sentenced_Pretria_sentenced_date2.index)):
    if Sentenced_Pretria_sentenced_date2.loc[i,'OFFENSE'] not in top_offense_sum.index[0:10]:
        Sentenced_Pretria_sentenced_date2.loc[i,'OFFENSE'] = 'OTHERS'

for i in range(len(Data_pretrial.index)):
    if Data_pretrial.loc[i,'OFFENSE'] not in top_offense_sum.index[0:10]:
        Data_pretrial.loc[i,'OFFENSE'] = 'OTHERS'
        
for i in range(len(Data_sentenced.index)):
    if Data_sentenced.loc[i,'OFFENSE'] not in top_offense_sum.index[0:10]:
        Data_sentenced.loc[i,'OFFENSE'] = 'OTHERS'

for i in range(len(Sentenced_Pretrial.index)):
    if Sentenced_Pretrial.loc[i,'OFFENSE'] not in top_offense_sum.index[0:10]:
        Sentenced_Pretrial.loc[i,'OFFENSE'] = 'OTHERS'
```

```python
Sentenced_Pretria_sentenced_date2
```

```python
#Find average sentenced date by race
Sentenced_average_race1 = Sentenced_Pretria_sentenced_date2.groupby(['RACE','OFFENSE'], as_index=False).SENTENCE_DAYS.mean()
Sentenced_average_race2 = Sentenced_Pretria_sentenced_date2.groupby(['RACE','OFFENSE'], as_index=False).BOND_AMOUNT.mean()
Sentenced_average_race3 = Sentenced_Pretria_sentenced_date2.groupby(['RACE','OFFENSE'], as_index=False).IDENTIFIER.count()
Sentenced_average_race4 = pd.merge(Sentenced_average_race1, Sentenced_average_race2, how='inner')
Sentenced_average_race = pd.merge(Sentenced_average_race4, Sentenced_average_race3, how='inner')
Sentenced_average_race
```

```python
Sentenced_average_gender1 = Sentenced_Pretria_sentenced_date2.groupby(['GENDER','OFFENSE'], as_index=False).SENTENCE_DAYS.mean()
Sentenced_average_gender2 = Sentenced_Pretria_sentenced_date2.groupby(['GENDER','OFFENSE'], as_index=False).BOND_AMOUNT.mean()
Sentenced_average_gender3 = Sentenced_Pretria_sentenced_date2.groupby(['GENDER','OFFENSE'], as_index=False).IDENTIFIER.count()
Sentenced_average_gender4 =  pd.merge(Sentenced_average_gender1, Sentenced_average_gender2, how='inner')
Sentenced_average_gender =  pd.merge(Sentenced_average_gender4, Sentenced_average_gender3, how='inner')
Sentenced_average_gender
```

```python
#Sum all data by Race and Gender
Data_pretrial_sum1 = Data_pretrial.groupby(['RACE','GENDER'])
Data_pretrial_sum1_df  = Data_pretrial_sum1.size()
Data_pretrial_sum1_df = Data_pretrial_sum1_df.to_frame()

Data_sentenced_sum1 = Data_sentenced.groupby(['RACE','GENDER'])
Data_sentenced_sum1_df  = Data_sentenced_sum1.size()
Data_sentenced_sum1_df = Data_sentenced_sum1_df.to_frame()

Sentenced_Pretrial_sum1 = Sentenced_Pretrial.groupby(['RACE','GENDER'])
Sentenced_Pretrial_sum1_df  = Sentenced_Pretrial_sum1.size()
Sentenced_Pretrial_sum1_df = Sentenced_Pretrial_sum1_df.to_frame()
```

```python
#Sum all data by Race and Offense
Data_pretrial_sum2 = Data_pretrial.groupby(['RACE','OFFENSE'])
Data_pretrial_sum2_df  = Data_pretrial_sum2.size()
Data_pretrial_sum2_df = Data_pretrial_sum2_df.to_frame()

Data_sentenced_sum2 = Data_sentenced.groupby(['RACE','OFFENSE'])
Data_sentenced_sum2_df  = Data_sentenced_sum2.size()
Data_sentenced_sum2_df = Data_sentenced_sum2_df.to_frame()

Sentenced_Pretrial_sum2 = Sentenced_Pretrial.groupby(['RACE','OFFENSE'])
Sentenced_Pretrial_sum2_df  = Sentenced_Pretrial_sum2.size()
Sentenced_Pretrial_sum2_df = Sentenced_Pretrial_sum2_df.to_frame()
```

```python
#Sum all data by Gender and Offense
Data_pretrial_sum3 = Data_pretrial.groupby(['GENDER','OFFENSE'])
Data_pretrial_sum3_df  = Data_pretrial_sum3.size()
Data_pretrial_sum3_df = Data_pretrial_sum3_df.to_frame()

Data_sentenced_sum3 = Data_sentenced.groupby(['GENDER','OFFENSE'])
Data_sentenced_sum3_df  = Data_sentenced_sum3.size()
Data_sentenced_sum3_df = Data_sentenced_sum3_df.to_frame()

Sentenced_Pretrial_sum3 = Sentenced_Pretrial.groupby(['GENDER','OFFENSE'])
Sentenced_Pretrial_sum3_df  = Sentenced_Pretrial_sum3.size()
Sentenced_Pretrial_sum3_df = Sentenced_Pretrial_sum3_df.to_frame()
```

```python
# Export as csv file
Data_pretrial.to_csv('Pretrial.csv')
Data_pretrial_sum1_df.to_csv('Pretrial_Sum_Race_Gender.csv')
Data_pretrial_sum2_df.to_csv('Pretrial_Sum_Race_Offense.csv')
Data_pretrial_sum3_df.to_csv('Pretrial_Sum_Gender_Offense.csv')
```

```python
# Export as csv file
Data_sentenced.to_csv('Sentenced.csv')
Data_sentenced_sum1_df.to_csv('Sentenced_Sum_Race_Gender.csv')
Data_sentenced_sum2_df.to_csv('Sentenced_Sum_Race_Offense.csv')
Data_sentenced_sum3_df.to_csv('Sentenced_Sum_Gender_Offense.csv')
```

```python
# Export as csv file
Sentenced_Pretria_sentenced_date2.to_csv('Sentenced_Pretrial.csv')
Sentenced_Pretrial_sum1_df.to_csv('Sentenced_Pretrial_Sum_Race_Gender.csv')
Sentenced_Pretrial_sum2_df.to_csv('Sentenced_Pretrial_Sum_Race_Offense.csv')
Sentenced_Pretrial_sum3_df.to_csv('Sentenced_Pretrial_Sum_Gender_Offense.csv')
```
